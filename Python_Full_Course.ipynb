```python
""" List """
list = [7,9,"harry"]
print(list[0] ,list[1] , list[2], list[0:2])
```

```python
""" List Methods Example """

# Original list
l1 = [1, 8, 7, 2, 21, 15]

# 1. sort() → Sorts the list in ascending order
l1.sort()
print("After sort:", l1)      # [1, 2, 7, 8, 15, 21]

# Reset list
l1 = [1, 8, 7, 2, 21, 15]

# 2. reverse() → Reverses the list
l1.reverse()
print("After reverse:", l1)   # [15, 21, 2, 7, 8, 1]

# 3. append() → Adds element at the end
l1.append(8)
print("After append(8):", l1)

# 4. insert() → Inserts element at given index
l1.insert(3, 8)
print("After insert(3, 8):", l1)

# 5. pop() → Removes element at given index
l1.pop(2)
print("After pop(2):", l1)

# 6. remove() → Removes the first occurrence of a value
l1.remove(21)
print("After remove(21):", l1)
```

```python
# ✅ Mutable (Can be changed after creation)
# You can modify the object in place.
# Examples:
# List
# Dictionary
# Set
# Bytearray
# Example (Mutable)
# l = [1, 2, 3]
# l[0] = 99
# print(l)   # [99, 2, 3]
# ✔ The list changed
# ✔ Same memory address




# ❌ Immutable (Cannot be changed after creation)
# You cannot modify the object once created.
# Any change creates a new object.
# Examples:
# int
# float
# string
# tuple
# bool
# frozenset
# Example (Immutable)
# s = "hello"
# s = s + " world"
# print(s)
# The original string "hello" never changed.
# A new string "hello world" was created.
```

```python
""" Sets """
#set is a collection of non-repetitive elements

s = set() # no repetition allowed!
s.add(1)
s.add(2)
s.add(1)
s.add(1)

print(s)
```

```python
# PROPERTIES OF SETS
# 1. Sets are unordered => Element’s order doesn’t matter
# 2. Sets are unindexed => Cannot access elements by index
# 3. There is no way to change items in sets.
# 4. Sets cannot contain duplicate values.
```

```python
""" Operations on Set """
s = {1,8,5,3,2,7,5,7,7,9}
print(s)

# len
print(len(s))

# remove
s.remove(8)
print(s)

# pop
s.pop()
print(s)

# clear
a = {32,4,5,65,6,5,5,5,5,5,242,4,24,345,3,46}
a.clear()
print(a)

# union
print(s.union({1,2,3,4,5},{8,6,5,4,3,2,3}))

# or operator also makes union
a = {1, 2, 3}
b = {3, 4, 5}
print(a | b)            # union
print(a.union(b))       # same


# intersection
print(s.intersection({1,2,3,4,5},{8,6,5,4,3,2,3}))


#  and opeator also makes intersection
a = {1, 2, 3}
b = {3, 4, 5}

print(a & b)                  # intersection
print(a.intersection(b))      # same
```

```python
# 4. Write a program to find whether a given number is prime or not.
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    limit = int(n**0.5) + 1               # square root
    for i in range(3, limit, 2):
        if n % i == 0:
            return False

    return True

num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
# 7. Write a python function to remove a given word from a list ad strip it at the same
# time
def remove_word(lst, word):
    result = []
    for item in lst:
        cleaned = item.strip()     # strip spaces
        if cleaned != word:        # skip the word to remove
            result.append(cleaned)
    return result

# Example usage
words = [" apple ", "banana", " mango ", "apple"]
print(remove_word(words, "apple"))
```

```python
import random

youStr = input("Enter your choice (s/w/g): ")

if youStr not in ['s', 'w', 'g']:
    print("Enter Valid Choice")
    exit()

youDict = {"s": 1, "w": -1, "g": 0}
reverseDict = {1: "Snake", -1: "Water", 0: "Gun"}

you = youDict[youStr]
computer = random.choice([1, -1, 0])

if computer == -1 and you == 1:
    result = "You Win!"
elif computer == -1 and you == 0:
    result = "You Win!"
elif computer == 1 and you == -1:
    result = "You Lose!"
elif computer == 1 and you == 0:
    result = "You Win!"
elif computer == 0 and you == 1:
    result = "You Lose!"
elif computer == 0 and you == -1:
    result = "You Lose!"
else:
    result = "Match Draw!"

print(result)
print("You chose:", reverseDict[you])
print("Computer chose:", reverseDict[computer])
```

```python
"""TYPES OF INHERITANCE """
# • Single inheritance
# • Multiple inheritance
# • Multilevel inheritance


""" Multiple inheritance """
class Employee:
  company="GOOGLE"
  name = "Ram"
  def show(self):
    print(f"The name of the employee is {self.name} and is woking in {self.company}")

class Coder:
  language = "Python"
  def languageShow(self):
    print(f"Language used by employee coder programmer is {self.language}")

class Programmer(Employee,Coder):
  salary = 120000
  def info(self):
    print(f"His salary is {self.salary}")

a = Employee()
b = Programmer()

a.show()
b.show()
b.info()
b.languageShow()
```

```python
""" Super Method """
# super() method is used to access the methods of a super class in the derived class.

class Employee:
  def __init__(self):
    super().__init__()
    print("Constructor of Employee")
  a=1
class Programmer(Employee):
  def __init__(self):
    super().__init__()
    print("Constructor of Programmer")
  b=1
class Manager(Programmer):
  def __init__(self):
    super().__init__()
    print("Constructor of Manager")
  c=3
o=Employee()
print(o.a)

o1=Programmer()
print(o1.a)
print(o1.b)

o2=Manager()
print(o2.a)
print(o2.b)
print(o2.c)
```

```python
""" Class Method """
# A class method is a method which is bound to the class and not the object of the class.
class Employee:
  a=1
  @classmethod
  def show(self):
    print(f"The Class Attribute of a is {self.a}")

e = Employee()
e.a=45
e.show()        # this shows 45 but we want to show 1 as we said class attrivute there

# So for that we use decorator @classmethod
```

```python
""" Property Decorators """
# Use case is when in a form I want to enter full name
# but the system needs separate first + last name.

class Employee:
  a = 1                  # class attribute, shared by all objects

  @classmethod
  def show(self):
    print(f"The Class Attribute of a is {self.a}")

  @property
  def name(self):
    return self.ename

  @name.setter
  def name(self, value):
    self.fname = value.split(" ")[0]
    self.lname = value.split(" ")[1]
    self.ename = value        # store original full name

e = Employee()
e.a = 45               # modifies INSTANCE attribute, not class attribute
e.show()

e.name = "Atharv Kulkarni"    # setter runs here
print(e.name)                 # getter runs here



# 🔍 Why use @property?
# ✔ Cleaner syntax
# Use obj.age instead of obj.get_age().
# ✔ Validation
# You can check values before assigning.
# ✔ Encapsulation
# Prevents direct access while keeping code readable.
# ✔ Backward compatibility
# You can convert attributes -> methods without breaking old code.
```

```python
# 🔥 FINAL SUMMARY
# ✔ Getter →          Method that reads a value
# ✔ Setter →          Method that writes/updates a value
# ✔ @property →       Makes getter work like an attribute  obj.value instead of obj.get_value()

# ✔ @attr.setter →

# Allows writing to that attribute:
# obj.value = something
```

```python
""" Operator Overloading """
# Operators in Python can be overloaded using dunder methods.
# These methods are called when a given operator is used on the objects.
# Operators in Python can be overloaded using the following methods:
# p1+p2     # p1.__add__(p2)
# p1-p2     # p1.__sub__(p2)
# p1*p2     # p1.__mul__(p2)
# p1/p2     # p1.__truediv__(p2)
# p1//p2    # p1.__floordiv__(p2)

# Other dunder/magic methods in Python:
# str__() # used to set what gets displayed upon calling str(obj)

# __len__() # used to set what gets displayed upon calling.__len__() or
# len(obj)
```

```python
class Number:
  def __init__(self,n):
    self.n=n
  def __add__(self,num):    #overloading of operator
    return self.n+num.n
  def __truediv__(self,num):    #overloading of operator
    return self.n/num.n


n=Number(1)
m=Number(2)

print(n+m)
print(n/m)
```

```python
# 1. Create a class (2-D vector) and use it to create another class representing a 3-D
# vector.

class TwoDVector:
    def __init__(self, i, j):
        self.i = i
        self.j = j

    def show(self):
        print(f"The vector is {self.i}i + {self.j}j ")


class ThreeDVector(TwoDVector):
    def __init__(self, i, j, k):
        super().__init__(i, j)
        self.k = k

    def show(self):
        print(f"The vector is {self.i}i + {self.j}j + {self.k}k")

a = TwoDVector(1, 2)
a.show()
b = ThreeDVector(5, 2, 3)
b.show()
```

```python
# 4. Write a class ‘Complex’ to represent complex numbers, along with overloaded
# operators ‘+’ and ‘*’ which adds and multiplies them.
class Complex:
    def __init__(self, r, i):
        self.r = r
        self.i = i

    def __add__(self, c2):
        return Complex(self.r + c2.r, self.i + c2.i)

    def __mul__(self, c2):
        real_part = self.r * c2.r - self.i * c2.i
        imag_part = self.r * c2.i + self.i * c2.r
        return Complex(real_part, imag_part)

    def __str__(self):                            #converts complex soln in comp language to  i and real format
        return f"{self.r} + {self.i}i"
# <__main__.Complex object at 0x7b38d0627d40>
# <__main__.Complex object at 0x7b38d0627d40>


c1 = Complex(1, 2)
c2 = Complex(3, 4)
print(c1 + c2)
print(c1*c2)
```

```python
# 5. Write a class vector representing a vector of n dimensions. Overload the + and *
# operator which calculates the sum and the dot(.) product of them.
class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __add__(self, other):
        result = Vector(self.x + other.x, self.y + other.y, self.z + other.z)
        return result

    def __mul__(self, other):
        result = self.x * other.x + self.y * other.y + self.z * other.z
        return result

    def __str__(self):
        return f"Vector({self.x}, {self.y}, {self.z})"

v1 = Vector(1, 2, 3)
v2 = Vector(4, 5, 6)
v3 = Vector(7, 8, 9)

print(v1 + v2)  # Output: Vector(5, 7, 9)
print(v1 * v2)  # Output: 32

print(v1 + v3)  # Output: Vector(8, 10, 12)
print(v1 * v3)  # Output: 50
```
